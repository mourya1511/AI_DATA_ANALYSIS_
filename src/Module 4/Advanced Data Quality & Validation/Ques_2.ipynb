```python
# Part 1: Detecting Data Drift in AI/ML Models
# Objective: Understand data drift, how it affects machine learning models, and techniques tomonitor it.

# Task 1: Understanding Data Drift: Study a historical dataset used in training a simple linear regression model and
# compare it with recent unseen data to detect drift.

# Task 2: Monitoring Distribution Changes: Write the code to identify features that exhibit statistical distribution differences.

# Task 3: Visualizing Data Drift: Use visualization techniques to illustrate data drift.

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# Simulating Historical and Recent Data
np.random.seed(42)

# Historical Data (Training Data)
historical_data = pd.DataFrame({
    'feature': np.random.normal(5, 2, 100),  # Normally distributed data
    'target': np.random.normal(10, 3, 100)
})

# Recent Data (Unseen Data)
recent_data = pd.DataFrame({
    'feature': np.random.normal(10, 2, 100),  # Shifted distribution (different mean)
    'target': np.random.normal(10, 3, 100)
})

# Train a Linear Regression Model on Historical Data
X_train = historical_data[['feature']]
y_train = historical_data['target']
model = LinearRegression()
model.fit(X_train, y_train)

# Predictions on Historical Data
y_train_pred = model.predict(X_train)
historical_mse = mean_squared_error(y_train, y_train_pred)

# Predictions on Recent Data
X_test = recent_data[['feature']]
y_test = recent_data['target']
y_test_pred = model.predict(X_test)
recent_mse = mean_squared_error(y_test, y_test_pred)

# Output the MSE for both datasets
print(f"Historical MSE: {historical_mse}")
print(f"Recent MSE: {recent_mse}")

# Visualization: Feature Distribution Comparison
plt.figure(figsize=(10, 6))
plt.hist(historical_data['feature'], bins=30, alpha=0.5, label='Historical Data')
plt.hist(recent_data['feature'], bins=30, alpha=0.5, label='Recent Data')
plt.legend()
plt.title("Feature Distribution Comparison (Historical vs Recent Data)")
plt.xlabel('Feature Value')
plt.ylabel('Frequency')
plt.show()


```
