```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error
from scipy.stats import ks_2samp
import seaborn as sns

# Sample Data (Replace this with your actual data)
# For illustration purposes, we assume 'historical_data' and 'recent_data' are DataFrames containing the data
# with a 'feature' column for the variable you're monitoring

# Example data for demonstration:
historical_data = pd.DataFrame({'feature': np.random.normal(0, 1, 1000)})
recent_data = pd.DataFrame({'feature': np.random.normal(0.5, 1.2, 1000)})

# Task 1: Calculate Mean Squared Error (MSE) for both datasets
historical_mse = mean_squared_error(historical_data['feature'], np.zeros_like(historical_data['feature']))
recent_mse = mean_squared_error(recent_data['feature'], np.zeros_like(recent_data['feature']))

# Task 2: Root Mean Squared Error (RMSE)
historical_rmse = np.sqrt(historical_mse)
recent_rmse = np.sqrt(recent_mse)

# Output the MSE and RMSE
print(f"Historical MSE: {historical_mse}")
print(f"Recent MSE: {recent_mse}")
print(f"Historical RMSE: {historical_rmse}")
print(f"Recent RMSE: {recent_rmse}")

# Task 3: Feature Distribution Comparison (Visualization)

plt.figure(figsize=(12, 8))

# Histograms for feature distribution comparison
plt.subplot(2, 2, 1)
plt.hist(historical_data['feature'], bins=30, alpha=0.5, label='Historical Data')
plt.hist(recent_data['feature'], bins=30, alpha=0.5, label='Recent Data')
plt.legend()
plt.title("Feature Distribution Comparison (Historical vs Recent Data)")
plt.xlabel('Feature Value')
plt.ylabel('Frequency')

# Boxplots for distribution comparison
plt.subplot(2, 2, 2)
sns.boxplot(data=[historical_data['feature'], recent_data['feature']], notch=True, patch_artist=True)
plt.xticks([0, 1], ['Historical Data', 'Recent Data'])
plt.title("Boxplot Comparison of Feature Distributions")

# Density plots for a smooth distribution comparison
plt.subplot(2, 2, 3)
sns.kdeplot(historical_data['feature'], label='Historical Data', shade=True)
sns.kdeplot(recent_data['feature'], label='Recent Data', shade=True)
plt.title("Density Plot Comparison of Feature Distributions")
plt.xlabel('Feature Value')
plt.ylabel('Density')

# KS Test: Kolmogorov-Smirnov Test for statistical distribution comparison
ks_statistic, ks_p_value = ks_2samp(historical_data['feature'], recent_data['feature'])
print(f"KS Test Statistic: {ks_statistic}")
print(f"KS Test p-value: {ks_p_value}")

# If p-value < 0.05, it indicates that the distributions are significantly different
if ks_p_value < 0.05:
    print("Data Drift Detected!")
else:
    print("No Significant Drift Detected.")

plt.tight_layout()
plt.show()
```
