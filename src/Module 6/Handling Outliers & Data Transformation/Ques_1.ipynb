```python
# Question: Visualizing Outliers Using Boxplots
# Description: Use a boxplot to visually detect outliers in a dataset.


import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

# Generate sample data with outliers
np.random.seed(10)
data = np.random.normal(loc=50, scale=10, size=100)
data = np.append(data, [120, 130, 140])  # Add outliers

# Create a DataFrame
df = pd.DataFrame({'Values': data})

# Plot boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(y='Values', data=df)
plt.title("Boxplot to Visualize Outliers")
plt.ylabel("Values")
plt.show()
```
