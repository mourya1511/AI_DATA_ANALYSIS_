```python
# Question: Power Transformation (Box-Cox) Method
# Description: Use Box-Cox transformation to stabilize variance and make the data more normal.


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import boxcox
import seaborn as sns

# Generate sample skewed data
np.random.seed(10)
data = np.random.exponential(scale=2, size=1000)

# Create a DataFrame
df = pd.DataFrame({'Values': data})

# Before Box-Cox transformation: Visualize with a histogram
plt.figure(figsize=(8, 5))
sns.histplot(df['Values'], kde=True, color='blue')
plt.title("Histogram Before Box-Cox Transformation")
plt.show()

# Apply Box-Cox transformation
transformed_data, lambda_opt = boxcox(df['Values'])

# Create a DataFrame with the transformed values
df_transformed = pd.DataFrame({'Transformed_Values': transformed_data})

# After Box-Cox transformation: Visualize with a histogram
plt.figure(figsize=(8, 5))
sns.histplot(df_transformed['Transformed_Values'], kde=True, color='red')
plt.title("Histogram After Box-Cox Transformation")
plt.show()

# Print the optimal lambda value
print(f"Optimal Lambda: {lambda_opt}")
```
