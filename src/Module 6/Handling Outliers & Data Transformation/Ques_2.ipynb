```python
# Question: Winsorization of Data
# Description: Apply Winsorization to a given dataset to handle outliers.

import numpy as np
import pandas as pd
from scipy.stats.mstats import winsorize
import matplotlib.pyplot as plt
import seaborn as sns

# Generate sample data with outliers
np.random.seed(10)
data = np.random.normal(loc=50, scale=10, size=100)
data = np.append(data, [120, 130, 140])  # Add outliers

# Create a DataFrame
df = pd.DataFrame({'Values': data})

# Before Winsorization: Visualize with a boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(y='Values', data=df)
plt.title("Boxplot Before Winsorization")
plt.show()

# Apply Winsorization to limit extreme values to the 5th and 95th percentiles
winsorized_data = winsorize(df['Values'], limits=[0.05, 0.05])

# Create a DataFrame with the Winsorized values
df_winsorized = pd.DataFrame({'Winsorized_Values': winsorized_data})

# After Winsorization: Visualize with a boxplot
plt.figure(figsize=(8, 5))
sns.boxplot(y='Winsorized_Values', data=df_winsorized)
plt.title("Boxplot After Winsorization")
plt.show()
```
