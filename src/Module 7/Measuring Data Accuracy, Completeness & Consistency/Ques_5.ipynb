## Validate Data Completeness for Critical Fields

**Description**: In `transaction_data.csv` , calculate the missing percentage of critical fields: "transaction_id", "amount", and "date", and visualize them.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Simulate some transaction data with missing values
data = {
    'transaction_id': [101, 102, 103, 104, np.nan, 106, 107],
    'amount': [100.0, np.nan, 150.5, 200.0, 250.0, np.nan, 300.0],
    'date': ['2023-01-01', '2023-01-02', np.nan, '2023-01-04', '2023-01-05', '2023-01-06', np.nan]
}

df = pd.DataFrame(data)

# Step 2: List of critical fields
critical_fields = ['transaction_id', 'amount', 'date']

# Step 3: Calculate missing data percentage
missing_data = {}
for field in critical_fields:
    missing_percentage = df[field].isnull().mean() * 100
    missing_data[field] = missing_percentage

# Step 4: Print missing percentages
for field, missing_percentage in missing_data.items():
    print(f"Missing percentage for '{field}': {missing_percentage:.2f}%")

# Step 5: Visualize missing data percentages
plt.figure(figsize=(8, 6))
plt.bar(missing_data.keys(), missing_data.values(), color='skyblue')
plt.title("Missing Data Percentage for Critical Fields", fontsize=14)
plt.ylabel("Missing Data Percentage (%)", fontsize=12)
plt.xlabel("Fields", fontsize=12)
plt.tight_layout()
plt.show()
```
