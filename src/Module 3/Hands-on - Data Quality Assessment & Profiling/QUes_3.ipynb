```python
# Part 1: Load a Dataset & Check Missing Values

# Task 1: Iris Dataset
# - Load the dataset using Pandas.
# - Check for missing values in the entire dataset.








# Part 2: Identify Duplicates & Inconsistencies

# Task 2: Inconsistent Entries in a Sample Dataset
# - Assume you have a dataset with a 'Gender' column. Identify inconsistent entries like 'M', 'Male', or 'male'.









# Part 3: Generate a Data Quality Report

# Task 3: Iris Dataset Summary
# - Generate basic descriptive statistics for the Iris dataset.


import pandas as pd
from sklearn.datasets import load_iris

def load_iris_dataset():
    """Loads the Iris dataset and returns a DataFrame."""
    iris = load_iris(as_frame=True)
    df = iris.frame
    df['species'] = iris.target_names[iris.target]  # Add species names
    print("‚úÖ Iris dataset loaded successfully.")
    return df

def check_missing_values(df):
    """Returns missing value count for each column."""
    return df.isnull().sum()

def check_gender_inconsistencies(df, column_name='Gender'):
    """Returns unique values in 'Gender' column to spot inconsistencies."""
    if column_name in df.columns:
        return df[column_name].unique()
    else:
        return f"‚ö†Ô∏è Column '{column_name}' not found."

def generate_statistics(df):
    """Generates basic descriptive statistics for a dataset."""
    return df.describe(include='all')

# --------- MAIN EXECUTION ---------

# --- Part 1: Iris Dataset - Missing Values ---
iris_df = load_iris_dataset()
missing_values = check_missing_values(iris_df)
print("\nüîç Missing Values in Iris Dataset:\n", missing_values)

# --- Part 2: Sample Dataset - Gender Inconsistencies ---
# Simulated sample dataset with 'Gender' column inconsistencies
sample_data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'Gender': ['Male', 'F', 'female', 'M', 'MALE']
}
sample_df = pd.DataFrame(sample_data)

gender_inconsistencies = check_gender_inconsistencies(sample_df)
print("\nüöª Unique values in 'Gender' column (check for inconsistencies):\n", gender_inconsistencies)

# --- Part 3: Data Quality Report - Iris Dataset Summary ---
stats = generate_statistics(iris_df)
print("\nüìà Basic Descriptive Statistics for Iris Dataset:\n", stats)

```
