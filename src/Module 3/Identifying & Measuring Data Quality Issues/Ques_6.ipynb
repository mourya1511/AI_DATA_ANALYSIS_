```python
# Evaluating Data Distribution using Randomly Generated Data

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import skew, kurtosis, mode

# Set seed for reproducibility
np.random.seed(42)

# Generate random data - Normal distribution
data = np.random.normal(loc=50, scale=10, size=1000)

# Create a DataFrame
df = pd.DataFrame(data, columns=['Value'])

# Plot the histogram
plt.figure(figsize=(10, 6))
sns.histplot(df['Value'], bins=30, kde=True, color='skyblue')
plt.title('Histogram of Generated Data')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.axvline(df['Value'].mean(), color='red', linestyle='--', linewidth=1.5, label='Mean')
plt.axvline(df['Value'].median(), color='green', linestyle='--', linewidth=1.5, label='Median')
plt.legend()
plt.show()

# Descriptive Statistics
mean_value = df['Value'].mean()
median_value = df['Value'].median()

# Handle mode extraction safely
mode_result = mode(df['Value'], keepdims=True)  # 'keepdims' ensures compatibility across versions
mode_value = mode_result.mode[0] if mode_result.mode.size > 0 else np.nan

std_dev = df['Value'].std()
skewness = skew(df['Value'])
kurt = kurtosis(df['Value'])

# Print Descriptive Statistics
print("Descriptive Statistics:")
print(f"Mean: {mean_value:.2f}")
print(f"Median: {median_value:.2f}")
print(f"Mode: {mode_value:.2f}")
print(f"Standard Deviation: {std_dev:.2f}")
print(f"Skewness: {skewness:.2f}")
print(f"Kurtosis: {kurt:.2f}")
```
