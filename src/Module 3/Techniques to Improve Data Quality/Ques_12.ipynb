```python
# Activity 2: Dealing with Duplicates & Redundancy

# Task A: Identifying Duplicate Records
import pandas as pd

# Create a sample DataFrame with some duplicate data
data = {
    'EmployeeID': [1, 2, 3, 4, 5, 6, 3],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Charlie'],
    'Age': [25, 30, 35, 40, 45, 50, 35],
    'Salary': [50000, 60000, 70000, 80000, 90000, 100000, 70000]
}

df = pd.DataFrame(data)

# Task 7: Identify Complete Duplicates
# Find complete duplicates (rows where all columns are the same)
duplicates_complete = df[df.duplicated(keep=False)]

print("Complete Duplicates:")
print(duplicates_complete)

# Task 8: Identify Duplicates Based on Specific Columns
# For example, check for duplicates based on the 'EmployeeID' and 'Salary' columns
duplicates_specific_columns = df[df.duplicated(subset=['EmployeeID', 'Salary'], keep=False)]

print("\nDuplicates Based on Specific Columns (EmployeeID and Salary):")
print(duplicates_specific_columns)

# Task 9: Count Duplicate Rows
# Count the number of complete duplicate rows
duplicate_count = df.duplicated(keep=False).sum()

# Print the count of duplicate rows
print("\nNumber of Duplicate Rows:")
print(duplicate_count)
# 7. Identify Complete Duplicates:
# - Load a dataset and identify duplicated rows.
# - Use Pandas to detect duplicates.






# 8. Identify Duplicates based on Specific Columns:
# - Check for duplicates in specified columns.







# 9. Count Duplicate Rows:
# - Calculate and print the number of duplicate rows.





```

```python
# Task B: Deduplication Techniques

# 10. Remove Complete Duplicates:
# - Drop duplicate rows and keep only the first occurrence.






# 11. Subset Deduplication:
# - Remove duplicates based on a subset of columns.






# 12. Keep Last Occurrence:
# - Drop duplicates but keep the last occurrence in the dataset.




import pandas as pd

# Create a sample DataFrame with some duplicate data
data = {
    'EmployeeID': [1, 2, 3, 4, 5, 6, 3, 2],
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Eva', 'Frank', 'Charlie', 'Bob'],
    'Age': [25, 30, 35, 40, 45, 50, 35, 30],
    'Salary': [50000, 60000, 70000, 80000, 90000, 100000, 70000, 60000]
}

df = pd.DataFrame(data)

# Task 10: Remove Complete Duplicates
# Drop complete duplicate rows and keep only the first occurrence
df_no_duplicates = df.drop_duplicates(keep='first')

print("DataFrame after removing complete duplicates (keeping the first occurrence):")
print(df_no_duplicates)

# Task 11: Subset Deduplication
# Remove duplicates based on a subset of columns, e.g., 'EmployeeID' and 'Salary'
df_no_duplicates_subset = df.drop_duplicates(subset=['EmployeeID', 'Salary'], keep='first')

print("\nDataFrame after removing duplicates based on 'EmployeeID' and 'Salary' (keeping the first occurrence):")
print(df_no_duplicates_subset)

# Task 12: Keep Last Occurrence
# Drop duplicates but keep the last occurrence of each duplicate row
df_no_duplicates_last = df.drop_duplicates(keep='last')

print("\nDataFrame after removing duplicates but keeping the last occurrence:")
print(df_no_duplicates_last)
```
