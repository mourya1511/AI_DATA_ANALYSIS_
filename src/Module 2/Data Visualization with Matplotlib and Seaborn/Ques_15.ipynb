```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the 'iris' dataset
iris = sns.load_dataset('iris')

# Exclude non-numeric columns (like 'species')
iris_numeric = iris.select_dtypes(include=['float64', 'int64'])

# Compute the correlation matrix for numeric columns only
correlation_matrix = iris_numeric.corr()

# Create a heatmap with advanced styling
plt.figure(figsize=(8, 6))  # Set figure size
sns.heatmap(
    correlation_matrix, 
    annot=True,        # Annotate with correlation values
    cmap='coolwarm',   # Color map (you can change this to other options like 'viridis', 'Blues', etc.)
    linewidths=0.5,    # Line width between cells
    fmt='.2f',         # Format the annotations to two decimal places
    cbar_kws={'label': 'Correlation Coefficient'},  # Color bar label
    square=True,       # Make the heatmap square
    vmax=1,            # Set max value for the colormap
    vmin=-1            # Set min value for the colormap
)
                                    
                                    # Customize the plot
plt.title('Correlation Matrix Heatmap of Iris Dataset', fontsize=16)
plt.show()
                                       # Customize the plot


```
