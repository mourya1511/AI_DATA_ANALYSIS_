```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data
heights = [160, 170, 165, 180, 175]
weights = [55, 65, 60, 80, 76]

# Create a DataFrame
data = pd.DataFrame({'Height (cm)': heights, 'Weight (kg)': weights})

# Create scatter plot
sns.scatterplot(x='Height (cm)', y='Weight (kg)', data=data, color='purple', s=100)

# Add title
plt.title('Height vs Weight')

# Display plot
plt.show()
```
