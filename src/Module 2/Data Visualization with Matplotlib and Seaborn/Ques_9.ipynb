```python
import matplotlib.pyplot as plt

# Define data

# Bar plot data
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [150, 200, 300, 250]

# Scatter plot data
heights = [160, 170, 165, 180, 175]
weights = [55, 65, 60, 80, 76]

# Create subplots (1 row, 2 columns)
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Bar Plot: Product Sales
axes[0].bar(products, sales, color='orange')
axes[0].set_title('Product Sales')
axes[0].set_xlabel('Product')
axes[0].set_ylabel('Sales')
axes[0].grid(axis='y', linestyle='--', alpha=0.7)

# Scatter Plot: Height vs Weight
axes[1].scatter(heights, weights, color='blue')
axes[1].set_title('Height vs Weight')
axes[1].set_xlabel('Height (cm)')
axes[1].set_ylabel('Weight (kg)')
axes[1].grid(True)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show plots
plt.show()

```
