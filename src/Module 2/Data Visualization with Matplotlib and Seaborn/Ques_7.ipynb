```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data
heights = [160, 170, 165, 180, 175]
weights = [55, 65, 60, 80, 76]
sizes = [10, 15, 13, 18, 12]

# Create DataFrame
data = pd.DataFrame({
    'Height (cm)': heights,
        'Weight (kg)': weights,
            'Size': sizes
            })

            # Create scatter plot with size
sns.scatterplot(
    x='Height (cm)',
    y='Weight (kg)',
    size='Size',
    data=data,
    legend=False,  # Hide size legend (optional)
    sizes=(50, 300),  # Control min/max size range for better visibility
    color='teal'
)

                                        # Customize plot
plt.title('Enhanced Height vs Weight Scatter Plot', fontsize=14)
plt.xlabel('Height (cm)', fontsize=12)
plt.ylabel('Weight (kg)', fontsize=12)

                                        # Show plot
plt.show()
```
