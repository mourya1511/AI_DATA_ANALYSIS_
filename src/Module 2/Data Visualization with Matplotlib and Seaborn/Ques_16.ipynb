```python
import matplotlib.pyplot as plt

# Data for plots
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [150, 200, 300, 250, 350]

heights = [160, 170, 165, 180, 175]
weights = [55, 65, 60, 80, 76]

products = ['Product A', 'Product B', 'Product C', 'Product D']
product_sales = [150, 200, 300, 250]

# Create a figure with subplots (1 row, 3 columns)
fig, axes = plt.subplots(1, 3, figsize=(15, 5))

# Line Plot: Sales Data
axes[0].plot(months, sales, color='green', marker='o', linestyle='-', linewidth=2, markersize=8)
axes[0].set_title('Monthly Sales')
axes[0].set_xlabel('Month')
axes[0].set_ylabel('Sales ($)')
axes[0].grid(True)

# Scatter Plot: Height vs Weight
axes[1].scatter(heights, weights, color='blue')
axes[1].set_title('Height vs Weight')
axes[1].set_xlabel('Height (cm)')
axes[1].set_ylabel('Weight (kg)')
axes[1].grid(True)

# Bar Plot: Product Sales
axes[2].bar(products, product_sales, color='orange')
axes[2].set_title('Product Sales')
axes[2].set_xlabel('Product')
axes[2].set_ylabel('Sales')
axes[2].grid(True, axis='y', linestyle='--', alpha=0.7)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show combined plot
plt.show()
```
