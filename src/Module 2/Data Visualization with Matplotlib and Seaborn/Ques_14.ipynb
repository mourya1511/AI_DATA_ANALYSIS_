```python
import matplotlib.pyplot as plt

# Data (Months vs Sales)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May']
sales = [150, 200, 300, 250, 350]

# Create a line plot with customized style
plt.plot(months, sales, color='green', marker='o', linestyle='-', linewidth=2, markersize=8, label='Sales')

# Customize title, labels, and grid
plt.title('Monthly Sales Data', fontsize=16, fontweight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Sales ($)', fontsize=12)

# Add a grid and customize it
plt.grid(True, linestyle='--', linewidth=0.7, alpha=0.7)

# Display the legend
plt.legend()

# Show the plot
plt.show()
```
