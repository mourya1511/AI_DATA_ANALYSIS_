```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample data
products = ['Product A', 'Product B', 'Product C', 'Product D']
sales = [150, 200, 300, 250]

# Create a DataFrame
data = pd.DataFrame({'Product': products, 'Sales': sales})

# Create bar plot
sns.barplot(x='Product', y='Sales', data=data, palette='Blues')

# Add title and labels
plt.title('Sales Comparison by Product')
plt.xlabel('Product')
plt.ylabel('Sales')

# Display the plo
plt.show()
```
